4.3.1 Задайте массив случайных значений из интервала(0; 1).
Рассчитайте средние и медианные значения для массива, сравните
результаты, какие выводы можно сделать о значениях?
Постройте точечную диаграмму рассения полученного ряда.


```python
from random import randint
import statistics as s
from matplotlib import pyplot as plt

arr = []
arr2 = []
for i in range(100):
    f = [randint(0, 1) for i in range(100)]
    arr.append(s.mean(f)) #среднее значение
    arr2.append(s.median(f)) #медианное значение 

plt.grid()
plt.scatter(arr,arr2) #зависимость среднего от медианнного 
plt.show()



```

4.3.2 Дана функция 
Создать массив из 10 значений
функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой
половины массива и построить графики для основного массива –
линейный и для среза – точечный


```python
import math as np 
from random import randint
import statistics as s
from matplotlib import pyplot as plt
import numpy as np

arr = [(np.sqrt(1+np.exp(np.sqrt(x))+np.cos(x**2)))/(np.fabs(1-np.sin(x)**3)) + np.log1p(np.fabs(x*2)) for x in range(1,10)]
print (arr)

plt.grid()
plt.plot(arr[:5],marker = "p")
plt.plot(arr)
plt.show()



```

4.3.3 
Дана функция |cos(𝑥𝑒(cos(𝑥)+ln(𝑥+1)))| Построить график на
интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь

```python
import math as m 
from random import randint
import statistics as s
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(0,10,1)
y = np.abs(np.cos(x*np.exp((np.cos(x)+np.log1p(x+1)))))

plt.grid()
plt.plot(x,y,c = "b")
plt.fill_between(x,y)

area = np.trapz(y)
print(area)
```
